```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 데이터셋 로드
food_data = pd.read_csv('./foodData4_24.csv')
food_data = food_data.head(5)

# 음식의 소화 후 남는 칼로리 계산 함수 정의
def calculate_digestible_calories(food_data):
    # 탄수화물, 지방, 단백질에 대한 흡수율 설정
    carbohydrate_absorption_rate = 0.08
    fat_absorption_rate = 0.03
    protein_absorption_rate = 0.30
    
    # 각 음식에 대해 소화 후 남는 칼로리 계산
    food_data['Digestible_Calories'] = (food_data['CarbohydrateContent'] * carbohydrate_absorption_rate * 4) \
                                       + (food_data['FatContent'] * fat_absorption_rate * 9) \
                                       + (food_data['ProteinContent'] * protein_absorption_rate * 4)

# 음식 데이터에 대해 소화 후 남는 칼로리 계산 함수 호출
calculate_digestible_calories(food_data)
```

```python
def calculate_digestible_calories(carbohydrate, fat, protein):
    # 탄수화물은 소화하는데 8%가 소모됨
    digestible_carbohydrate_calories = carbohydrate * 0.92 * 4
    # 지방은 소화하는데 3%가 소모됨
    digestible_fat_calories = fat * 0.97 * 9
    # 단백질은 소화하는데 30%가 소모됨
    digestible_protein_calories = protein * 0.7 * 4
    
    # 소화 후 남는 칼로리 계산
    digestible_calories = digestible_carbohydrate_calories + digestible_fat_calories + digestible_protein_calories
    
    return digestible_calories

# 음식의 영양소 정보 입력
carbohydrate = 37.1  # 탄수화물 함량 (g)
fat = 2.5  # 지방 함량 (g)
protein = 3.2  # 단백질 함량 (g)
total_calories = 170.9  # 총 칼로리 (kcal)

# 소화 후 남는 칼로리 계산
digestible_calories = calculate_digestible_calories(carbohydrate, fat, protein)

# 결과 출력
print("소화 후 남는 칼로리:", digestible_calories)
```
