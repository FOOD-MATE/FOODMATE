```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def weighted_nutrition(df, nutrition):
    percentile = 0.6
    m = df[nutrition].quantile(percentile)
    C = df['Calories'].mean()
    v = df[nutrition]
    R = df['Calories']

    return (((v / (v+m)) * R) + ((m / (v+m)) * C)) / R

def recommend_similar_food(user_similar_food, food):
    # 사용자의 선호도를 DataFrame으로 변환
    user_preferences = pd.DataFrame(user_similar_food, columns=['CarbohydrateContent', 'ProteinContent', 'FatContent'])

    # 음식과 사용자의 선호도 간의 코사인 유사도 계산
    similarities = cosine_similarity(user_preferences, food[['Carbohydrate_weighted_rate', 'Protein_weighted_rate', 'Fat_weighted_rate']])

    # 코사인 유사도가 가장 높은 음식 추천
    recommended_food_index = similarities.argmax()
    recommended_food = food.iloc[recommended_food_index]['FoodName']
    
    return recommended_food

# 음식 데이터 로드
food = pd.read_csv('./foodData4_24.csv')

# 각 영양소에 대한 가중평균 계산
food['Carbohydrate_weighted_rate'] = weighted_nutrition(food, 'CarbohydrateContent')
food['Protein_weighted_rate'] = weighted_nutrition(food, 'ProteinContent')
food['Fat_weighted_rate'] = weighted_nutrition(food, 'FatContent')

# 사용자의 각 영양소에 대한 가중평균값 계산
user_similar = []
similar_keywords_food_weighted(food)

# 음식 추천
recommended_food = recommend_similar_food(user_similar, food)
print("추천 음식:", recommended_food)
```
